#NDCG

```python
def get_logs(scores):
  return [scores[i]/math.log2(i+2) for i in range(len(scores))]

def ndcg_score(scores, true_scores):
  
  dcg = sum(get_logs(scores))

  idcg = sum(get_logs(true_scores))

  ndcg = dcg/idcg

  return ndcg
```

```python
y_true = [[0, 1, 0]]
y_pred = [[0, 1, 0]]
```

```python
from sklearn.metrics import ndcg_score as ns
```

```python
ndcg_score(y_pred[0], y_true[0])
```

```python
ns(y_true, y_pred)
```

# Top N Accuracy

```python
def topNacc(preds, n, trues):

  got_right = []
  for pred in preds[:n]:
    if pred in trues:
      got_right.append(pred)
  
  acc = len(got_right)/n
  return acc
```

```python
def acc_scores(all_preds, n, all_trues):

  scores = []
  for query_preds, true in zip(all_preds, all_trues):
    scores.append(topNacc(query_preds, n, true))
  
  #return
```

```python
scores = [3, 1, 0, 0, 0]
trues = [1, 2, 3, 4, 5]

topNacc(scores, 1, trues)
```

# Mean Reciprocal Rank

```python
def get_rr_score(query):
  preds, trues = query[0], query[1]
  for i in range(len(preds)):
    if preds[i] in trues:
      return 1/(i+1)
    if i == (len(preds)-1):
      return 0
```

```python
query_1 = [[0,1,0],[1,2,3]]
query_2 = [[4,6,7],[4,5,6]]
queries = [query_1, query_2]
```

```python
for i in queries:
  print(get_rr_score(i))
```

```python
def mean_rr(queries):
  return (sum([get_rr_score(query) for query in queries]))/len(queries)
```

```python
mean_rr(queries)
```
